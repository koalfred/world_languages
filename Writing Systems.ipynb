```python
import pygal
from pygal.style import Style
```

## Writing systems

There are many writing systems used around the world to write different languages. The code below generates a map showing the distribution of different writing systems worldwide.

```python
writing_systems_style = Style(colors = (
    '#940c9e', '#38c76c', '#187ce7', '#f5940a', '#f6b009', '#bf072b', '#0514fa', '#fbe68f', '#ce0a7d', '#00d7ff',
    '#f59dcd', '#f6ff00', '#f708a8', '#875b0f'
))

writing_systems = pygal.maps.world.World(style=writing_systems_style)

writing_systems.title = 'Writing systems predominantly used in different places around the world'

writing_systems.add('Latin', {
    'an', 'al', 'ao', 'ar', 'at', 'au', 'az', 'ba', 'be', 'bf', 'bi', 'bj',
    'bn', 'bo', 'br', 'bw', 'bz', 'ca', 'cd', 'cf', 'cg', 'ch', 'ci', 'cl',
    'cm', 'co', 'cr', 'cu', 'cv', 'cz', 'de', 'dk', 'do', 'ec', 'ee', 'es',
    'fi', 'fr', 'ga', 'gb', 'gf', 'gh', 'gl', 'gm', 'gn', 'gq', 'gt', 'gu',
    'gw', 'gy', 'hn', 'hr', 'ht', 'hu', 'id', 'ie', 'is', 'it', 'jm', 'ke',
    'li', 'lr', 'ls', 'lt', 'lu', 'lv', 'mc', 'md', 'me', 'mg', 'ml', 'mt',
    'mu', 'mw', 'mx', 'my', 'mz', 'na', 'ne', 'ng', 'ni', 'nl', 'no', 'nz',
    'pa', 'pe', 'pg', 'ph', 'pl', 'pr', 'pt', 'py', 're', 'ro', 'rw', 'sc',
    'se', 'sg', 'sh', 'si', 'sk', 'sl', 'sm', 'se', 'sn', 'so', 'sr', 'st',
    'sv', 'sz', 'tg', 'tl', 'tm', 'tr', 'tz', 'ug', 'us', 'uy', 'uz', 'va',
    've', 'vn', 'yt', 'za', 'zm', 'zw'
})

writing_systems.add('Arabic/Perso-Arabic', {
    'ae', 'af', 'bh', 'dj', 'dz', 'eg', 'eh', 'iq', 'ir', 'jo', 'kw', 'lb',
    'ly', 'ma', 'mr', 'om', 'pk', 'ps', 'sa', 'sd', 'sy', 'td', 'tn', 'ye'
})

writing_systems.add('Cyrillic', {
    'bg', 'by', 'kg', 'kz', 'mk', 'mn', 'rs', 'ru', 'tj', 'ua'
})

writing_systems.add('Armenian', {
    'am'
})

writing_systems.add('North Indic', {
    'bd', 'bt', 'in', 'np'
})

writing_systems.add('Hanzi', {
    'cn', 'hk', 'mo', 'tw'
})

writing_systems.add('Greek', {
    'cy', 'gr'
})

writing_systems.add('Ethiopic', {
    'er', 'et'
})

writing_systems.add('Georgian', {
    'ge'
})

writing_systems.add('Hebrew', {
    'il'
})

writing_systems.add('Hiragana, Katakana and Kanji', {
    'jp'
})

writing_systems.add('South Indic', {
    'kh', 'la', 'lk', 'mm', 'th'
})

writing_systems.add('Hangul', {
    'kp', 'kr'
})

writing_systems.add('Thaana', {
    'mv'
})

writing_systems.render_to_file('writing_system_maps/writing_systems.svg')
```
